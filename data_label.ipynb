```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data/LUNA25_Public_Training_Development_Data.csv')
```

```python
df
```

```python
df.groupby('SeriesInstanceUID', as_index=False)['label'].nunique().sort_values('label', ascending=False)
```

```python
X = df.drop('label', axis=1)
y = df['label']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1)
```

```python
pd.concat([X_train, y_train], axis=1).to_csv('data/train.csv', index=False)
```

```python
pd.concat([X_val, y_val], axis=1).to_csv('data/valid.csv', index=False)
```
